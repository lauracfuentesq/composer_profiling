# 2. Tokenizing lieder score data

```python
import json
from pathlib import Path

file=open('../data preprocessing/corpus_conversion_midi.json')
scores_mapping=json.load(file)

scores_paths=[]
scores_paths_json=[]

for score in scores_mapping:    
    scores_paths.append(score['out'])

    json_path=score['out'].replace('.mid','.json')
    json_path=json_path.replace('Lieder/scores','author-profiling-in-symbolic-music/lieder_tokens')    
    scores_paths_json.append(json_path)

# convert the list to Path objects
scores_paths_path = [Path(score) for score in scores_paths]
```

```python
tokenizer_BPE_REMI.tokenize_dataset(files_paths=scores_paths_path,out_dir='lieder_tokens')
```

```python
import pandas as pd

# gather json files with token ids paths
json_file_list=[]

for score in scores_paths_json:
    file=open(score)
    json_file_list.append(json.load(file))

# create a pandas dataframe with the token ids per score
lieder_tokens_df=pd.DataFrame(json_file_list)

lieder_tokens_df['n_TokSequence_objs']=lieder_tokens_df['ids'].apply(lambda token_ids : len(token_ids))

lieder_tokens_df['len_TokSequence_ids']=lieder_tokens_df['ids'].apply(lambda lists: [len(inner_list) for inner_list in lists])

# compute additional variables with the mido library
from mido import MidiFile

scores_tracks_names=[]

for score in scores_paths:

    score_tracks_names=[]
    for track in MidiFile(score).tracks:
        score_tracks_names.append(track.name)

    scores_tracks_names.append(score_tracks_names)

lieder_tokens_df['scores_track_names']=scores_tracks_names

lieder_tokens_df['n_tracks']=lieder_tokens_df['scores_track_names'].apply(lambda x: len(x))

lieder_tokens_df['n_tracks_equal_nsequence?']=lieder_tokens_df['n_TokSequence_objs']==lieder_tokens_df['n_tracks']
```

```python
##==== compute track names mapping =====##

unique_track_names_list=[]

for track_names in lieder_tokens_df['scores_track_names']:    
    unique_track_names=list(set(track_names))

    for track_name in unique_track_names:

        if track_name not in unique_track_names_list:
            unique_track_names_list.append(track_name)


clean_track_names=['piano',
 'voice',
 'voice',
 'piano',
 'other',
 'voice',
 'other',
 'voice',
 'voice',
 'other',
 'piano',
 'voice',
 'other',
 'voice',
 'voice',
 'other',
 'other',
 'voice',
 'other',
 'other',
 'voice',
 'piano',
 'voice',
 'voice',
 'voice',
 'voice',
 'voice',
 'voice',
 'voice',
 'voice',
 'voice',
 'other',
 'voice',
 'voice',
 'voice',
 'voice',
 'voice',
 'voice',
 'voice',
 'voice',
 'voice',
 'piano',
 'other',
 'other',
 'voice',
 'voice',
 'voice',
 'voice',
 'voice',
 'voice',
 'voice',
 'voice',
 'voice',
 'voice',
 'voice',
 'voice',
 'voice',
 'voice',
 'voice',
 'piano',
 'voice',
 'other']

track_names_mapping={}
for key,value in zip(unique_track_names_list,clean_track_names):
    track_names_mapping[key]=value

names_mapping_df=pd.DataFrame.from_dict(track_names_mapping, orient='index', columns=['track_name_clean']).reset_index()
names_mapping_df.rename(columns={'index':'track_name_raw'},inplace=True)

#names_mapping_df.to_csv('track_names_mapping_df.csv',index=False)
```

```python
track_names_mapping_df=pd.read_csv('../dataframes/track_names_mapping_df.csv')

track_names_mapping={}
for key,value in zip(track_names_mapping_df['track_name_raw'],track_names_mapping_df['track_name_clean']):
    track_names_mapping[key]=value

lieder_tokens_df['scores_track_names_mapping']=lieder_tokens_df['scores_track_names'].apply(lambda names_list: [track_names_mapping[name] for name in names_list])

lieder_tokens_df['len_TokSequence_ids'] = lieder_tokens_df['len_TokSequence_ids'].apply(ast.literal_eval)

lieder_tokens_df['scores_paths_formatted']=lieder_tokens_df['scores_paths'].apply(lambda x: x.split('/')[3:6])
lieder_tokens_df['scores_paths_formatted']=lieder_tokens_df['scores_paths_formatted'].apply(lambda x: '/'.join(x))

temp=lieder_tokens_df[lieder_tokens_df['n_tracks_equal_nsequence?']==True][['scores_track_names_mapping','len_TokSequence_ids']]
##======== only tokens df ======##

# Explode the DataFrame
tracks_tokens_length = temp.explode(['scores_track_names_mapping', 'len_TokSequence_ids'])

tracks_tokens_length.reset_index(drop=True,inplace=True)

tracks_tokens_length['len_TokSequence_ids']=tracks_tokens_length['len_TokSequence_ids'].astype(int)

# Group by 'scores_track_names_mapping' and describe 'len_TokSequence_ids'
grouped_description = tracks_tokens_length.groupby('scores_track_names_mapping')['len_TokSequence_ids'].describe()
```
